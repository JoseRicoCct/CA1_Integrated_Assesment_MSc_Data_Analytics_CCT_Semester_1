## 1. Looking up Population Estimates (Persons in April), PEA01.20231004T131011.csv:

```python
# import pandas data set
import pandas as pd
```

```python
# Loading the dataset:
df1 = pd.read_csv('..\\Datasets_data.cso.ie\\PEA01.20231004T131011.csv')
```

```python
# Displaying first 5 rows:
df1.head()
```

```python
# Looking up distinct values within the df:
def df1_distinct(df1):
    for column in df1:
        unique_values = df1[column].unique()
        print(f"Unique values for {column}:")
        print(unique_values)
        print("\n")

# Call the function with your DataFrame as an argument
df1_distinct(df1)
```

```python
# Getting df info
df1.info()
```

```python
# Describing the df:
df1.describe()
```

```python
# Counting the NaN values:
df1.isnull().sum()
```

```python
# Display the rows with null values
df1_NaN = df1[df1.isnull().any(axis=1)]
print("Rows with null values:")
df1_NaN.head(10)
```

```python
#Function to count the NaN values per category:
def df1_NaN_Count(df1_NaN):
    for column in df1_NaN:
        counts = df1_NaN[column].value_counts()
        print(f"Count {column}:")
        print(counts)
        print("\n")


# Call the function with your DataFrame as an argument
df1_NaN_Count(df1_NaN)
```

Conclusion:

This dataset has 114 missing values for category Age Group 0 - 4 years. It has 5994 entries.
